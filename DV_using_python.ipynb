```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/Large_SaaS_Churn_Dummy_Data__100K_rows_.csv')

# Quick check
print(data.head())
print(data.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plan_churn = data.groupby('plan_type')['churned'].mean().reset_index()

sns.barplot(x='plan_type', y='churned', data=plan_churn)
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Plan Type')
plt.show()
```

```python
acct_churn = data.groupby('has_account_manager')['churned'].mean().reset_index()

sns.barplot(x='has_account_manager', y='churned', data=acct_churn)
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Account Manager')
plt.show()
```

```python
heatmap_data = data.pivot_table(index='tenure_months', columns='plan_type', values='churned', aggfunc='mean')

sns.heatmap(heatmap_data, cmap='coolwarm')
plt.title('Heatmap of Churn by Plan Type and Tenure')
plt.show()
```
